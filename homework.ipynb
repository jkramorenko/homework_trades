```python
import pandas as pd

data_url = "https://github.com/jkramorenko/homework_trades/blob/main/data.xls?raw=True"
requests = pd.read_excel(data_url, sheet_name="Заявки клиентов")
trades = pd.read_excel(data_url, sheet_name="Сделки клиентов")
all_market = pd.read_excel(data_url, sheet_name="Сделки всего рынка")

pd.options.display.float_format = '{:,.8f}'.format
pd.options.display.max_rows = 100
pd.options.display.min_rows = 100
```

```python
# <1.1 Объем сделок рынка по инструменту>
all_market.groupby(['Инструмент','Дата']).aggregate({'Объем':'sum'})
```

```python
# <1.2 Объем сделок клиентов по инструменту>
trades.groupby(['Инструмент','Дата']).aggregate({'Объем':'sum'})
```

```python
# <1.3 Средневзвешенная цена по инструменту>
all_market.groupby(['Инструмент','Дата']).apply(lambda group: group['Сумма'].sum() / group['Объем'].sum())
```

```python
# <1.4 Количество заявок клиентов по инструменту>
requests.groupby(['Инструмент','Дата']).aggregate({'Объем':'count'})
```

```python
# <1.5 Количество снятых заявок клиента по инструменту>
requests.loc[requests['Объем'] == requests['Остаток объема']].groupby(['Инструмент','Дата']).aggregate({'Объем':'count'})
```

```python
# <1.6 Количество сделок клиента с контрагентом>
res = trades.groupby(['Инструмент','Дата']).aggregate({'Клиент-констрагент':'count'})
res[res['Клиент-констрагент']>0]
```

```python
# <2.1 Доля клиентов в объеме рынка>

total_sum = trades['Сумма'].sum()

trades.groupby(['Клиент']).apply(lambda group: 100 * group['Сумма'].sum() / total_sum)
```

```python
# <2.2 Сделки, цена которых отклоняется от средневзвешенной более чем на 1, 2 или 5% (выбрать порог самостоятельно по количеству сигналов)>
wavg = all_market.groupby(['Инструмент']).apply(lambda group: group['Сумма'].sum() / group['Объем'].sum())

def filter_avg(row):
    avg_value = wavg[row['Инструмент']]
    return (100 * abs(row['Цена'] - avg_value)/avg_value) > 5

all_market[all_market.apply(filter_avg, axis=1)]
```

```python
# <2.4 Максимальное количество единовременно снятых клиентом заявок>
requests[['Клиент','Время снятия заявки', 'Номер заявки']].rename(columns={'Номер заявки':'Count'}).groupby(['Клиент','Время снятия заявки']).count().sort_values(by='Count').groupby(level=0).tail(1)
```
